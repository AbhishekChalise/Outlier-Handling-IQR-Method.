```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
```

```python
# The data values
data = {
    'values': [1,2,10,11,100,12,31,44,55,66,77,44,55,45,65,78,98,32,34,54,67,87,97,64,56,67,86,45,66,76,77,5000]
}
df = pd.DataFrame(data)
df
```

 # Quartile 1 which is the 25%  

```python
Q1 = np.quantile(df['values'],q = 0.25)
print(Q1)
```

```python
# Third Quartile
Q3 = np.quantile(df['values'],q = 0.75)
print(Q3)
```

```python
# So the Range of values is (39.0 and 76.5)
# Now lets find the upper bound and the lower bound
IQR = Q3 - Q1
print(IQR)

lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
print(lower_bound,upper_bound)
```

```python
s = df[ (df['values'] <= upper_bound) & (df['values'] >= lower_bound)]   
s
```

```python
# Before Removing The Outliers
sb.boxplot(df)
```

```python
# After Removing The Outliers
sb.boxplot(s)
```
